```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from math import floor, sin, cos
```

```python
plt.rcParams["figure.figsize"] = (20,5)
```

```python
def func(x):
    return sin(x) * cos(x)
```

```python
def quantize(y, step, mode = "midtread"):
    if mode == "midtread":  
        return floor(y / step + 0.5) * step
    else:
        return (floor(y/step) + 0.5) * step
```

```python
def calcStep(X, Y, n_levels):
    return (max(Y) - min(Y)) / (n_levels - 1)
```

```python
X = np.arange(1, 10, 0.5)
Y = [func(x) for x in X]
```

```python
n_levels = 24
```

```python
step = calcStep(X, Y, n_levels)
```

```python
Y_QUANT = [quantize(y, step, "midrise") for y in Y]
```

```python
ticks = [tick for tick in np.arange(min(Y), max(Y), step)]
fig, (ax1, ax2, ax3) = plt.subplots(1, 3)
ax1.plot(X,Y)

ax2.plot(X,Y)
ax2.step(X,Y_QUANT)
ax2.set_yticks(ticks)
ax2.grid(axis = 'y')

ax3.plot(X,Y)
ax3.plot(X,Y_QUANT)
ax3.set_yticks(ticks)
ax3.grid(axis = 'y')
```
